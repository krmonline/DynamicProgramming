# Min Coin
มีเหรียญ 1,2,5 บาท

จะถือเงินน้อยที่สุดกี่เหรียญ

```
#MinCoin
import pandas as pd
import numpy as np
```

```
def minCoin(x):
  if x < 0:
    return 0
  if x == 0:
    return 0
  if myDict[x] != 0:
    return myDict[x]
  else:
    c1 = x - 1
    cum1 = minCoin(c1) + 1 if c1 >= 0 else 100
    c2 = x - 2
    cum2 = minCoin(c2) + 1 if c2 >= 0 else 100
    c5 = x - 5
    cum5 = minCoin(c5) + 1 if c5 >= 0 else 100
  #print(x,cum1,cum2,cum5)
  myDict[x] = minSelect(cum1,cum2,cum5)
  return myDict[x]
def minSelect(x,y,z):
  return min(x,y,z)
```

# ร้านขนม
ร้านขายขนม ทำขนมมา 2 สูตร
สูตร 1  ใช้ Chocolate 1 unit , นม 3 unit ได้กำไร 6 บาท
สูตร 2  ใช้ Chocolate 1 unit , นม 2 unit ได้กำไร 5 บาท
หากมี นม 12 unit และ Chocolate 5 unit 
ทำยังไงให้ได้กำไรมากที่สุด

```
#Chocolate
#Formula 1 1 chocolate,3 milk = 6 bath
#Formula 2 1 chocolate,2 milk = 5 bath
#We have 5 chocolate and 12 milk
def maxBath(c,m):
  if c <= 0 or m <= 0:
    return 0
  if c >= 1 and m >= 3:
    c1 = c-1
    m1 = m-3
    cum1 = maxBath(c1,m1) + 6
  else:
    cum1 = 0

  if c >= 1 and m >= 2:
    c2 = c-1
    m2 = m-2
    cum2 = maxBath(c2,m2) + 5
  else:
    cum2 = 0
  
  myDict[c][m] = maxSelect(cum1,cum2)
  return myDict[c][m]

def maxSelect(x,y):
  return max(x,y)
```

```
c = 5
m = 12
myDict = np.zeros([c+1,m+1])
print(maxBath(5,12))
#print(myDict)
```

### น้ำมัน Desel 100 ลิตร ราคาทุนลิตรละ 12 บาทและน้ำมันปปาล์ม 120 ลิตร ราคาต้นทุนลิตรละ 8 บาท ถ้าผสมน้ำมันทั้งสองชนิด รวมกันให้มีจำนวนไม่น้อยกว่า 150 ลิตร และขายน้ำมันผสมนี้ในราคาลิตรละ 11 บาทให้ได้กำไรมากที่สุดแล้ว กำไรที่ได้เท่ากับเท่าไหร่

```
import numpy as np
def maxOilProfit(x,dl,pl):
  if x <= 0:
    return 0
  if myDict[x][dl][pl] != 0:
    return myDict[x][dl][pl]
  d = maxOilProfit(x-1,dl-1,pl) + 11 - 12 if dl > 0 else -100
  p = maxOilProfit(x-1,dl,pl-1) + 11 - 8 if pl  > 0 else -100
  myDict[x][dl][pl] = np.max([d,p])
  return myDict[x][dl][pl]
```

```
myDict = np.zeros([151,101,121])
maxOilProfit(150,100,120)
```

ต้องการซื้อรถตู้เพื่อทำรถขนส่งโดยมีงบประมาณ 100,000 ต่อเดือน และค่าบำรุงรักษาไม่เกิน 500 ต่อเดือน

รถตู้ A ราคา 10,000 ต่อเดือน มีค่าบำรุงรักษา 100 ต่อเดือนนั่งได้ 7 คน

รถตู้ B ราคา 20,000 ต่อเดือน มีคำบำรุงรักษา 75 ต่อเดือนนั่งได้ 15 คน

ควรที่จะซื้อรถตู้ชนิดละกี่คันเพื่อให้ได้จำนวนคนนั่งสูงที่สุด

```
import numpy as np
def maxPassengers(vanPrice,maPrice,numVA,numVB):
  #Scale 100,000 to 10 because memory limitation on colab
  vanPrice = int(vanPrice)
  maPrice = int(maPrice)
  if vanPrice <= 0  or maPrice <= 0:
    return 0
  if myDict[vanPrice][maPrice][numVA][numVB] != 0:
    return myDict[vanPrice][maPrice][numVA][numVB]
  VAPassengers = maxPassengers(vanPrice-1.0000,maPrice-100,numVA+1,numVB) + 7 if vanPrice >= 1.0000 and maPrice >= 100 else 0
  VBPassengers = maxPassengers(vanPrice-2.0000,maPrice-75,numVA,numVB+1) + 15  if vanPrice >= 2.0000 and maPrice >= 75 else 0
  myDict[vanPrice][maPrice][numVA][numVB] = np.max([VAPassengers,VBPassengers])
  return myDict[vanPrice][maPrice][numVA][numVB]
```

```
myDict = np.zeros([11,501,100,100])
maxPassengers(10,500,0,0)
```
